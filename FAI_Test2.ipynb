```python
# Import required libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset from seaborn
iris = sns.load_dataset('iris')

# Example 1: Using Matplotlib (plt)
plt.figure(figsize=(8, 6))
plt.scatter(iris['sepal_length'], iris['sepal_width'])
plt.title('Sepal Length vs Width')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()

# Example 2: Using Seaborn (sns)
plt.figure(figsize=(8, 6))
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species')
plt.title('Sepal Length vs Width by Species')
plt.show()
```

```python
# Import required libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the iris dataset into a DataFrame
iris = sns.load_dataset('iris')

# Example 1: Basic DataFrame operations and plotting
# Select specific columns
df = iris[['sepal_length', 'sepal_width', 'species']]

# Calculate mean sepal length by species
mean_by_species = df.groupby('species')['sepal_length'].mean()

# Plot using the DataFrame
plt.figure(figsize=(8, 6))
mean_by_species.plot(kind='bar')
plt.title('Average Sepal Length by Species')
plt.ylabel('Sepal Length')
plt.show()

# Example 2: DataFrame filtering and plotting
# Filter for specific species
setosa_df = df[df['species'] == 'setosa']
versicolor_df = df[df['species'] == 'versicolor']

# Create scatter plot using filtered DataFrames
plt.figure(figsize=(8, 6))
plt.scatter(setosa_df['sepal_length'], setosa_df['sepal_width'], label='Setosa')
plt.scatter(versicolor_df['sepal_length'], versicolor_df['sepal_width'], label='Versicolor')
plt.title('Sepal Length vs Width by Species')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.legend()
plt.show()

# Example 3: Using DataFrame with Seaborn
# Create new column in DataFrame
iris['sepal_ratio'] = iris['sepal_length'] / iris['sepal_width']

# Plot using the new column
plt.figure(figsize=(8, 6))
sns.boxplot(data=iris, x='species', y='sepal_ratio')
plt.title('Sepal Ratio Distribution by Species')
plt.show()

# Example 4: Multiple DataFrame operations
# Calculate summary statistics
summary_df = iris.groupby('species').agg({
    'sepal_length': ['mean', 'min', 'max'],
    'sepal_width': ['mean', 'min', 'max']
}).round(2)

# Plot multiple subplots using DataFrame
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

summary_df['sepal_length']['mean'].plot(kind='bar', ax=ax1)
ax1.set_title('Mean Sepal Length')

summary_df['sepal_width']['mean'].plot(kind='bar', ax=ax2)
ax2.set_title('Mean Sepal Width')

plt.tight_layout()
plt.show()
```
