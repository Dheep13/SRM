```python
# Poisson distribution example: Let's say the average number of events (e.g., calls to a call center) is 2 per hour.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as pltx
import scipy.stats as st
# import statsmodels.stats as sm

# Parameters for the Poisson distribution
lambda_poisson = 2  # average rate of occurrence (lambda)

# Possible outcomes (number of events)
x_poisson = np.arange(0, 10)

# Probability of getting exactly x events
probabilities_poisson = st.poisson.pmf(x_poisson, lambda_poisson)

# Plotting the Poisson distribution
plt.figure(figsize=(8, 6))
plt.bar(x_poisson, probabilities_poisson, color='lightgreen', edgecolor='black')

# Adding labels and title
plt.title('Poisson Distribution: Average of 2 Events per Hour', fontsize=14)
plt.xlabel('Number of Events', fontsize=12)
plt.ylabel('Probability P(X=x)', fontsize=12)

# Display the plot
plt.grid(True)
plt.show()

# Calculate the variance for the Poisson distribution
variance_poisson = lambda_poisson  # Variance in Poisson is equal to the rate (lambda)
variance_poisson
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as st

# Generate x values
x = np.arange(51)  # Assuming n=50, so x goes from 0 to 50

# Calculate CDF
cf = st.binom.cdf(x, 50, 11/15)

# Create the plot
plt.figure(figsize=(20, 15))
sns.set_style("whitegrid")
sns.set_palette("viridis")

# Create barplot with custom colors
bars = sns.barplot(x=x, y=cf)

# Customize colors
n_bars = len(bars.patches)
colors = plt.cm.viridis(np.linspace(0, 1, n_bars))
for bar, color in zip(bars.patches, colors):
    bar.set_color(color)

# Customize the plot
plt.title("Binomial Cumulative Distribution Function", fontsize=20)
plt.xlabel("Number of Successes", fontsize=16)
plt.ylabel("Cumulative Probability", fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Add a colorbar
sm = plt.cm.ScalarMappable(cmap="viridis", norm=plt.Normalize(vmin=0, vmax=50))
sm.set_array([])
cbar = plt.colorbar(sm)
cbar.set_label("Number of Successes", fontsize=14)

plt.tight_layout()
plt.show()
```
