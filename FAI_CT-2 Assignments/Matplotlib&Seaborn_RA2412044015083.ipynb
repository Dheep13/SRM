```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create sample data
np.random.seed(42)
data = {
    'Group': ['A'] * 30 + ['B'] * 30 + ['C'] * 30,
    'Values': np.concatenate([
        np.random.normal(100, 10, 30),
        np.random.normal(90, 15, 30),
        np.random.normal(110, 12, 30)
    ]),
    'Scores': np.random.uniform(60, 100, 90),
    'Category': np.random.choice(['High', 'Medium', 'Low'], 90)
}
df = pd.DataFrame(data)

# Create correlation matrix data
corr_data = pd.DataFrame(np.random.rand(5, 5), 
                        columns=['Var1', 'Var2', 'Var3', 'Var4', 'Var5'])

# Create multiple plots
plt.figure(figsize=(15, 10))

# Bar Plot
plt.subplot(2, 3, 1)
sns.barplot(data=df, x='Group', y='Values', hue='Group', legend=False)
plt.title('Bar Plot')
plt.xlabel('Groups')
plt.ylabel('Average Values')

# Box Plot
plt.subplot(2, 3, 2)
sns.boxplot(data=df, x='Group', y='Values', hue='Group', legend=False)
plt.title('Box Plot')
plt.xlabel('Groups')
plt.ylabel('Distribution of Values')

# Violin Plot
plt.subplot(2, 3, 3)
sns.violinplot(data=df, x='Group', y='Values', hue='Group', legend=False)
plt.title('Violin Plot')
plt.xlabel('Groups')
plt.ylabel('Distribution of Values')

# Scatter Plot
plt.subplot(2, 3, 4)
plt.scatter(df['Values'], df['Scores'], alpha=0.5)
plt.title('Scatter Plot')
plt.xlabel('Values')
plt.ylabel('Scores')

# Heatmap
plt.subplot(2, 3, 5)
sns.heatmap(corr_data.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')

# Distribution Plot
plt.subplot(2, 3, 6)
sns.histplot(data=df, x='Values', hue='Group', multiple="stack")
plt.title('Distribution Plot')
plt.xlabel('Values')
plt.ylabel('Count')

plt.tight_layout()
plt.show()

# Advanced Box Plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Group', y='Values', hue='Category')
plt.title('Box Plot by Group and Category', pad=20)
plt.xlabel('Groups', labelpad=10)
plt.ylabel('Values', labelpad=10)
plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# Advanced Heatmap
plt.figure(figsize=(8, 6))
mask = np.triu(np.ones_like(corr_data.corr()))
sns.heatmap(corr_data.corr(), 
            mask=mask,
            annot=True,
            cmap='viridis',
            fmt='.2f',
            square=True,
            linewidths=0.5,
            cbar_kws={"shrink": .5})
plt.title('Advanced Correlation Heatmap')
plt.tight_layout()
plt.show()

# Split Violin Plot
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='Group', y='Values', hue='Category',
               split=True, inner='quart')
plt.title('Split Violin Plot by Group and Category')
plt.xlabel('Groups')
plt.ylabel('Distribution of Values')
plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# Joint Plot
sns.jointplot(data=df, x='Values', y='Scores', 
             kind='hex', height=8, marginal_kws={'bins': 30})
plt.suptitle('Joint Distribution of Values and Scores', y=1.02)
plt.show()
```
