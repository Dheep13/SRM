```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some data
x = np.linspace(0, 10, 100)
y = np.sin(x)

print(x)
# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='sin(x)')
plt.title('Simple Sine Wave Plot')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate x values
x = np.linspace(-10, 10, 100)

# Define two quadratic equations
y1 = x**2 + 2*x + 1  # y = x^2 + 2x + 1
y2 = 2*x**2 - 4*x - 2  # y = 2x^2 - 4x - 2

# Create the plot
plt.figure(figsize=(5, 5))
plt.plot(x, y1, label='y = x^2 + 2x + 1', color='blue')
plt.plot(x, y2, label='y = 2x^2 - 4x - 2', color='red')

# Customize the plot
plt.title('Two Quadratic Equations')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

# Add x and y axis lines
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the bar graph
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 35, 14, 28, 19]

# Create the bar graph
plt.figure(figsize=(10, 6))
bars = plt.bar(categories, values, color='skyblue', edgecolor='navy')

# Customize the plot
plt.title('Simple Bar Graph')
plt.xlabel('Categories')
plt.ylabel('Values')

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height}',
             ha='center', va='bottom')

# Add a grid for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter

# Generate some random data
np.random.seed(42)  # for reproducibility
data = np.random.normal(100, 20, 1000)  # 1000 points with mean=100, std=20

# Round the data to integers
rounded_data = np.round(data).astype(int)

# Count the frequency of each integer
frequency_count = Counter(rounded_data)

# Create the histogram
plt.figure(figsize=(12, 6))
n, bins, patches = plt.hist(data, bins=30, edgecolor='black')

# Customize the plot
plt.title('Histogram of Normal Distribution (Focused on High Frequency Values)')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Add a grid for better readability
plt.grid(axis='y', alpha=0.75)

# Function to get exact frequency of a specific integer
def get_exact_frequency(value):
    return frequency_count[value]

# Specific integers to highlight
integers_to_highlight = [80, 100, 120]
colors = ['green', 'blue', 'purple']

for value, color in zip(integers_to_highlight, colors):
    freq = get_exact_frequency(value)
    plt.plot([value, value], [0, freq], color=color, linestyle='--', linewidth=2)
    plt.text(value, freq, f'Value: {value}\nFreq: {freq}', 
             color=color, ha='right', va='bottom')

# Focus the view on the central part of the distribution
mean = np.mean(data)
std = np.std(data)
plt.xlim(mean - 2*std, mean + 2*std)  # Show data within 2 standard deviations of the mean

# Show the plot
plt.tight_layout()
plt.show()

# Print the exact frequencies
print("Exact integer frequencies:")
for value in integers_to_highlight:
    print(f"Integer {value}: {get_exact_frequency(value)} occurrences")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some random data
np.random.seed(42)  # for reproducibility
data = np.random.normal(100, 20, 1000)  # 1000 points with mean=100, std=20

# Calculate bin range and width
data_min = np.min(data)
data_max = np.max(data)
num_bins = 30
bin_width = (data_max - data_min) / num_bins

print(f"Data range: {data_min:.2f} to {data_max:.2f}")
print(f"Number of bins: {num_bins}")
print(f"Calculated bin width: {bin_width:.2f}")

# Create custom bins
custom_bins = np.linspace(data_min, data_max, num_bins + 1)

# Create the histogram
plt.figure(figsize=(12, 6))
n, bins, patches = plt.hist(data, bins=custom_bins, edgecolor='black')

# Customize the plot
plt.title('Histogram of Normal Distribution with Custom Bins')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Add a grid for better readability
plt.grid(axis='y', alpha=0.75)

plt.tight_layout()
plt.show()

# Print information about the bins
print("\nBin information:")
for i in range(len(n)):
    print(f"Bin {i+1}: Range [{bins[i]:.2f}, {bins[i+1]:.2f}), Count: {n[i]}")

print(f"\nTotal data points: {sum(n)}")
print(f"Actual number of bins: {len(n)}")
```

```python
## Matplot lib vs Seaborn comparison

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Generate some data
np.random.seed(0)
data = np.random.randn(1000)

# Matplotlib histogram
plt.figure(figsize=(12, 5))
plt.subplot(121)
plt.hist(data, bins=30)
plt.title('Matplotlib Histogram')

# Seaborn histogram
plt.subplot(122)
sns.histplot(data, bins=30, kde=True)
plt.title('Seaborn Histogram')

plt.tight_layout()
plt.show()

# Seaborn additional plot types
plt.figure(figsize=(12, 5))
plt.subplot(121)
sns.boxplot(data=data)
plt.title('Seaborn Boxplot')

plt.subplot(122)
sns.violinplot(data=data)
plt.title('Seaborn Violinplot')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Example 1: List of lists to DataFrame
print("Example 1: List of lists to DataFrame")
list_data = [
    ['Alice', 25, 'New York'],
    ['Bob', 30, 'San Francisco'],
    ['Charlie', 35, 'Los Angeles'],
    ['David', 28, 'Chicago']
]

df_from_list = pd.DataFrame(list_data, columns=['Name', 'Age', 'City'])
print(df_from_list)
print("\n")

# Example 2: Dictionary to DataFrame
print("Example 2: Dictionary to DataFrame")
dict_data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 28],
    'City': ['New York', 'San Francisco', 'Los Angeles', 'Chicago']
}

df_from_dict = pd.DataFrame(dict_data)
print(df_from_dict)
print("\n")

# Example 3: List of dictionaries to DataFrame
print("Example 3: List of dictionaries to DataFrame")
list_of_dicts = [
    {'Name': 'Alice', 'Age': 25, 'City': 'New York'},
    {'Name': 'Bob', 'Age': 30, 'City': 'San Francisco'},
    {'Name': 'Charlie', 'Age': 35, 'City': 'Los Angeles'},
    {'Name': 'David', 'Age': 28, 'City': 'Chicago'}
]

df_from_list_of_dicts = pd.DataFrame(list_of_dicts)
print(df_from_list_of_dicts)
print("\n")

# Comparing the results
print("Are all DataFrames equal?")
print(df_from_list.equals(df_from_dict) and df_from_dict.equals(df_from_list_of_dicts))
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Create a dataset with outliers
np.random.seed(0)
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'] * 20,
    'Department': ['Sales', 'IT', 'Marketing', 'HR', 'Finance'] * 20,
    'Salary': np.random.randint(30000, 100000, 100),
}

df = pd.DataFrame(data)

# Add outliers
outliers = {
    'Name': ['Outlier1', 'Outlier2', 'Outlier3', 'Outlier4', 'Outlier5'],
    'Department': ['Sales', 'IT', 'Marketing', 'HR', 'Finance'],
    'Salary': [150000, 200000, 180000, 160000, 250000],
}

df = pd.concat([df, pd.DataFrame(outliers)], ignore_index=True)

# Create the salary boxplot
plt.figure(figsize=(15, 8))
ax = sns.boxplot(x='Department', y='Salary', data=df)

# Function to add statistical annotations
def add_statistical_annotations(ax, data, x, y):
    for i, dept in enumerate(data[x].unique()):
        dept_data = data[data[x] == dept][y]
        
        # Calculate statistics
        mean = dept_data.mean()
        median = dept_data.median()
        q1, q3 = dept_data.quantile([0.25, 0.75])
        iqr = q3 - q1
        
        # Add mean marker
        ax.plot(i, mean, marker='D', color='red', markersize=8)
        ax.text(i, mean, f'Mean\n{mean:.0f}', ha='center', va='bottom', color='red')
        
        # Add IQR annotation
        ax.text(i, q3, f'Q3: {q3:.0f}', ha='left', va='bottom')
        ax.text(i, q1, f'Q1: {q1:.0f}', ha='left', va='top')
        ax.text(i, median, f'IQR: {iqr:.0f}', ha='right', va='center')
        
        # Add whisker annotations
        whisker_low = max(dept_data.min(), q1 - 1.5 * iqr)
        whisker_high = min(dept_data.max(), q3 + 1.5 * iqr)
        ax.text(i, whisker_low, f'Lower\nWhisker', ha='left', va='top')
        ax.text(i, whisker_high, f'Upper\nWhisker', ha='left', va='bottom')

# Add annotations to the plot
add_statistical_annotations(ax, df, 'Department', 'Salary')

# Customize the plot
plt.title('Salary Distribution by Department (with Statistical Annotations)', fontsize=16)
plt.xlabel('Department', fontsize=12)
plt.ylabel('Salary', fontsize=12)

# Adjust layout and show the plot
plt.tight_layout()
plt.show()

# Print summary statistics for Salary
print("Salary Summary Statistics:")
print(df.groupby('Department')['Salary'].describe())

# Display information about the outliers
print("\nOutlier Information:")
print(df[df['Name'].str.startswith('Outlier')])
```

```python
import matplotlib.pyplot as plt

# Data for the chart
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [30, 25, 20, 25]

# Create a pie chart and make it a donut by setting 'wedgeprops'
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, wedgeprops={'linewidth': 3, 'edgecolor': 'white'})

# Draw a circle at the center to make it a donut chart
centre_circle = plt.Circle((0,0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Set the title
plt.title('Fruit Distribution - Donut Chart')

# Display the chart
plt.show()
```
