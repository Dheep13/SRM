```python
# Problem: Confidence Interval for Population Mean

from scipy import stats
import math

# Given values
n = 250
x_bar = 45000
sigma = 4000
confidence_level = 0.95


# Z-score for 95% confidence level
z_score = stats.norm.ppf(1 - (1 - confidence_level) / 2)

# Calculate the standard error
SE = sigma / math.sqrt(n)

# Calculate the margin of error
margin_of_error = z_score * SE

# Calculate the confidence interval
lower_bound = x_bar - margin_of_error
upper_bound = x_bar + margin_of_error

(lower_bound, upper_bound)
```

```python
# Problem: Confidence Interval for Population Proportion

from math import sqrt
from scipy import stats

# Given values for the problem
n_customers = 1000
satisfied_customers = 820
confidence_level = 0.99

# Step 1: Calculate the sample proportion (p)
p_hat = satisfied_customers / n_customers

# Step 2: Find the Z-score for 99% confidence level
z_score = stats.norm.ppf(1 - (1 - confidence_level) / 2)

# Step 3: Calculate the standard error (SE)
SE = sqrt((p_hat * (1 - p_hat)) / n_customers)

# Step 4: Calculate the margin of error (ME)
ME = z_score * SE

# Step 5: Calculate the confidence interval
lower_bound = p_hat - ME
upper_bound = p_hat + ME

(lower_bound, upper_bound)
```
